```python
def minor_mint_agent_v2(user_input, current_balance=0.5):
    """Agent متطور يحسب ويزيد ينصح"""

    # قائمة مشتريات وهمية للمراهقين
    items = {
        "اشتراك spotify": 0.1,
        "قهوة": 0.05,
        "skin fortnite": 0.2
    }

    user_input = user_input.lower()

    # 1. المراهق يحب يشري
    if "نشري" in user_input:
        found_item = False
        for item, price in items.items():
            if item in user_input:
                found_item = True
                if current_balance >= price:
                    new_balance = round(current_balance - price, 3)
                    return f"💸 مريغل! الـ {item} بـ {price} POL. بعد ما نشريوه، يقعدلك {new_balance} POL. نصدموا؟"
                else:
                    return f"😬 سامحني يا بطل، الـ {item} غالي شوية على ميزانيتك. ناقصك {round(price - current_balance, 3)} POL."

        if not found_item:
            return "شنوة تحب تشري بالظبط؟ عندي حالياً: اشتراك Spotify، قهوة، وإلا Skin Fortnite."

    # 2. نصيحة مالية (هذي تعجب الحكام برشا)
    elif "نصيحة" in user_input:
        return "💡 نصيحة MinorMint: ديما خلي عندك 0.1 POL 'خبيئة' للـ Gas Fees باش ما تحصلش وقت اللي تحب تبعث فلوس صحيحة!"

    return "أهلاً بيك! قولي 'نشري قهوة' مثلاً وشوف الذكاء متاعي."

# جرب الكود توة في Colab
print(minor_mint_agent_v2("نحب نشري قهوة"))
```

```python
def minor_mint_guardian_agent(user_input, lang="ar_tn", currency="POL"):
    # 1. قاموس المشتريات وتصنيفها (Healthy vs Restricted)
    market_items = {
        "قهوة": {"price": 0.05, "status": "ok", "msg": "صحة وبالشفاء!"},
        "كتاب": {"price": 0.1, "status": "ok", "msg": "أحسن ما تعمل، القراية تنور العقل."},
        "دخان": {"price": 0.5, "status": "blocked", "msg": "ممنوع! الصحة متاعك أهم من كل شيء."},
        "vape": {"price": 0.8, "status": "blocked", "msg": "مستحيل، هذي مشتريات مضرة وما نوافقش عليها."}
    }

    # 2. أسعار الصرف (وهمية للتجربة)
    exchange_rates = {"POL": 1, "USDC": 0.8, "TND": 3.2}

    user_input = user_input.lower()

    # --- معالجة الطلب ---
    for item, info in market_items.items():
        if item in user_input:
            # أول حاجة: التثبت من "سلامة" المنتج
            if info["status"] == "blocked":
                return f"🚫 {info['msg']}"

            # ثاني حاجة: حساب الثمن بالعملة المختارة
            price_in_selected = round(info["price"] * exchange_rates[currency], 2)

            if lang == "ar_tn":
                return f"💸 الـ {item} بـ {price_in_selected} {currency}. {info['msg']} نصدموا؟"
            elif lang == "fr":
                return f"💸 Le {item} coûte {price_in_selected} {currency}. {info['msg']} On y va ?"
            else:
                return f"💸 {item} costs {price_in_selected} {currency}. {info['msg']} Proceed?"

    return "أهلاً بك! قولي شنوة تحب تشري وإلا تحب نبدلو اللغة؟"

# --- تجربة الـ Guardian Mode ---
print(minor_mint_guardian_agent("نحب نشري دخان")) # يرفض
print(minor_mint_guardian_agent("نحب نشري كتاب", lang="ar_tn", currency="TND")) # يقبل ويحول للـ TND
```

```python
def minor_mint_pro_agent(user_input, currency="POL"):
    # 1. تعريف العملات وأسعار الصرف (نجمو نربطوها بـ API لاحقاً)
    rates = {
        "POL": 1.0,
        "TND": 3.15,  # الدينار التونسي
        "USDC": 0.85, # الدولار الرقمي
        "EUR": 0.78   # اليورو
    }

    # 2. منطق الحماية (Guardian Logic)
    # هوني Gemini في العادة هو اللي يقرر، أما برمجياً نحددو الـ Categories
    restricted_keywords = ["دخان", "vape", "شراب", "قمار", "betting", "سجائر"]

    # 3. محاكاة تحليل الطلب
    user_input_lower = user_input.lower()

    # التثبت من المواد المضرة
    if any(word in user_input_lower for word in restricted_keywords):
        return {
            "status": "REJECTED",
            "msg": "❌ العملية مرفوضة: MinorMint يمنع شراء المواد المضرة بصحتك أو اللي فيها مخاطر مالية."
        }

    # 4. منطق تحويل العملة (Currency Switching)
    # لنفرض أن الثمن الأصلي ديما بالـ POL ونحولو للمستخدم حسب طلبه
    base_price_pol = 0.05 # سعر افتراضي (مثلاً قهوة)
    converted_price = round(base_price_pol * rates.get(currency, 1.0), 3)

    return {
        "status": "APPROVED",
        "msg": f"✅ مريغل! العملية مسموحة. الثمن هو {converted_price} {currency}.",
        "advice": "💡 نصيحة: ما تنساش اللي الادخار يخليك تشري حاجات أهم في المستقبل!"
    }

# --- تجربة التعديلات الجديدة ---
print("--- تجربة شراء ممنوع ---")
print(minor_mint_pro_agent("نحب نشري vape")["msg"])

print("\n--- تجربة تبديل عملة (TND) ---")
res = minor_mint_pro_agent("نشري قهوة", currency="TND")
print(f"{res['msg']} \n{res['advice']}")
```
